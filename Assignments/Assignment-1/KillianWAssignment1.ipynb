```python
#Importing the needed libraries
import math
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
#PROBLEM 1 
#Part A
def VectorLength(x,y):
    return np.sqrt(x**2 + y**2)
#Part B
print(VectorLength(5,6))
print(VectorLength(-1,5))
#Part C
print("sin(1) = " + str(round(np.sin(1),3))) #Alternative way of formatting to 3 decimal places
```

```python
#PROBLEM 2
def IsPrime(n):  #Checks if a number n is prime
    if (n==1):   #Part B
        return False
    if (n<4):
        return True
    for k in range(2,math.floor(np.sqrt(n)+1)):
        if (n%k == 0):
            return False
    return True

#Part A
for i in range(1,11):     #Demonstrating that the function IsPrime works correctly for 10 examples
    if (IsPrime(i) == False):
        print(str(i) + " is not prime")
    else:
        print(str(i) + " is prime")

#Part E
primes_less_than_1000 = 0
for i in range(1,1001):
    if(IsPrime(i)==True):
        primes_less_than_1000 += 1
print("\nThe number of primes less than 1000 is " + str(primes_less_than_1000))

#Part F
def PrimeCount(N):
    primes_less_than_N = 0
    for i in range(1,N+1):
        if(IsPrime(i)==True):
            primes_less_than_N += 1
    return primes_less_than_N
print("\nUsing the PrimeCount function, the number of primes less than 1,000 is "+ str(PrimeCount(1000)))
print("Using the PrimeCount function, the number of primes less than 10,000 is "+ str(PrimeCount(10000)))

#Part G
x = np.arange(0,40,1)
Pi = np.zeros(40)
i=0
while i < np.size(Pi):
    Pi[i] = PrimeCount(i)
    i+=1
plt.plot(x, Pi, 'ro')
plt.xlabel("x")
plt.ylabel("Pi(x)")
plt.show()
```

```python
#PROBLEM 3
sample_data = np.loadtxt('PowerLawOrExponential.txt')
x_values = sample_data[:,0]
y_values = sample_data[:,1]
plt.plot(x_values,y_values)
plt.title("Ordinary Linear-Linear plot")
plt.grid(True,'both','both',linestyle='--')
plt.xlabel("X",fontsize='large')
plt.ylabel("Y",fontsize='large')
plt.show()
print("It is hard to determine from this graph whether our function is a power law or exponential.")
plt.plot(x_values,np.log(y_values),'pink')
plt.title("Linear - Log plot")
plt.grid(True,'both','both',linestyle='--')
plt.xlabel("x",fontsize='large')
plt.ylabel("ln(y)",fontsize='large')
plt.show()
print("This function seems to have power law behavior.")
plt.plot(np.log(x_values), np.log(y_values),'r')
plt.title("Log - Log plot")
plt.grid(True,'both','both',linestyle='--')
plt.xlabel("ln(x)",fontsize='large')
plt.ylabel("ln(y)",fontsize='large')
plt.show()
print("As the function is linear on a log-log scale, we conclude that the function has power law behavior.")
print("Additionally, as the slope of this graph is 2, we conclude that our data is sampled from the function f(x) = x^2.")
```
