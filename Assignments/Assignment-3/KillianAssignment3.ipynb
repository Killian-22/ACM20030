## Question 1: Linear least-square fitting
This questions asks you to write a function to find the best linear fit to a dataset and then apply that function to a few data sets. The normal equations for linear least-squares fitting are given by:  
  
$$
  \begin{align*}
    m \sum x_i^2 + c \sum x_i &= \sum x_i y_i \\
    m \sum x_i + cn &= \sum y_i
  \end{align*}
$$
  
Write these simultaneous equations in matrix for $A.x = b$, where $A$ is a $2\times2$ matrix and $x=\{m,c\}$ and $b$ are column vectors. Note that for a $2\times2$ matrix 
$$
A = \begin{pmatrix}
    a  & b \\
    c  & d 
\end{pmatrix}
$$
the inverse is given by
$$
A^{-1} = \frac{1}{ad-bc}\begin{pmatrix}
    d  & -b \\
    -c  & a 
\end{pmatrix}
$$

### Q1a.

Write a function that takes in two arrays of data for the x and y values and returns the values of $m$ and $c$ from for the linear least squares fit. Call the function `LinearLeastSquares`. You might want to use the function `np.dot(A,b)` which will take the dot product of a matrix, $A$ and a vector, $b$.

```python
def LinearLeastSquares(x_values,y_values):
    sum_x = np.sum(x_values)
    sum_y = np.sum(y_values)
    sum_x2 = np.sum((x_values)**2)
    sum_y2 = np.sum((y_values)**2)
    sum_xy = np.sum(x_values*y_values)
    n = len(x_values)
    v = np.matrix([[sum_xy],[sum_y]])
    inverse_matrix = (1/(sum_x2*n - (sum_x)**2))*np.matrix([[n,-sum_x],[-sum_x,sum_x2]])
    return inverse_matrix*v
```

## Question 2: second-order ODEs

Use the below second-order version of the (fixed-step) RK2 function for this question.

```python
def RK2(f, dx, x0, y0, imax):
    output = np.empty((imax, 3))
    i = 0
    xi = x0
    yi = y0
    while(i < imax):
        k1 = dx*f(xi, yi)
        k2 = dx*f(xi + dx, yi + k1)
        yi = yi + 0.5*(k1 + k2)
        xi += dx
        output[i, 0] = xi
        output[i, 1] = yi[0]
        output[i, 2] = yi[1]
        i += 1
    return output
```

### Q2a

Write the follow ODEs in first-order form and then solve them using the `RK2` function. Plot the numerical and analytic solution on the same plot (with a grid, labelling the axes and a legend).

$$
 y''(x) - y'(x) +\frac{1}{4} y(x) = -\sin(x)
$$
with $y(0) = 0, y'(0) = 73/100$. Find the numerical solution between $x=0$ and $x=9$.

This has the analytic solution:

$$
y(x) = \frac{1}{100} \left(-7 e^{x/2} x+64 e^{x/2}+48 \sin (x)-64 \cos (x)\right)
$$

For the numerical solution, start with a step size of $\Delta x = 0.1$. You may need to adjust this to get the numerical solution to agree with the analytic solution.

Our original equation is $$y''(x)-y'(x)+\frac{y(x)}{4} = -sin(x)$$
However, we will first find solutions to the homogeneous problem, namely
$$y''(x)-y'(x)+\frac{y(x)}{4} = 0$$
We let $y(x) = s_1(x)$ and $y'(x) = s_2(x)$
Then, we can use the homogeneous equation to write a system of ODEs:
$$\frac{d}{dx}\begin{bmatrix} s_1(x)\\s_2(x)\end{bmatrix}= \begin{bmatrix} 0 & 1\\\frac{-1}{4} & 1\end{bmatrix}\begin{bmatrix} s_1(x)\\s_2(x)\end{bmatrix}$$
The method of undetermined coefficients can be easily used here to find the particular solution $$y_p(x) = \frac{1}{100}(48\sin(x)-64\cos(x))$$
We can simply combine the solution of the linear system from the homogeneous equation with the particular solution given to produce an exact solution of the differential equation.

How small do you have to make the step size for the numerical solution to agree well with the analytic solution? What does this suggest about the equation?

```python
def System1(x,s):
    return np.array([s[1],(-s[0]/4 + s[1] - np.sin(x))])
num_solution_largestep = RK2(System1,0.1,0,[0,73/100],90)
num_solution_smallstep = RK2(System1,0.01,0,[0,73/100],900)
x = np.linspace(0,9,9000)
def exact_solution(x): #This is the analytic solution of the differential equation.
    return((1/100)*(-7*np.exp(x/2)*x + 64*np.exp(x/2) + 48*np.sin(x) - 64*np.cos(x)))
plt.scatter(num_solution_largestep[:,0],num_solution_largestep[:,1],color='green',label='step size=0.1')
plt.scatter(num_solution_smallstep[:,0],num_solution_smallstep[:,1],marker='.',color='pink',label='step size=0.01')
plt.xlabel('x')
plt.ylabel('y(x)')
plt.grid(True, 'both', 'both', linestyle='--')
plt.plot(x,exact_solution(x),color='red',label='exact solution',linewidth=2)
plt.legend()
plt.show()
```

Here we see that although the 0.1 step size numerical solution differs appreciably from the exact solution, the 0.01 step size numerical solution does a very good job approximating the analytic solution. This necessity of small step sizes indicates that the differential equation is "stiff".

## Question 3: chaos

(Q3a): Use SciPy's `solve_ivp` to solve the Lorenz system:

$$\begin{align}
\frac{dx}{dt} &= \sigma(y-x) \\
\frac{dy}{dt} &= x(\rho -z) - y   \\
\frac{dz}{dt} &= x y- \beta z
\end{align}$$

with $ \rho = 28$, $\sigma = 10$ and $\beta = 8/3$, and initial conditions $x(0) = 1, y(0) = 1, z(0) = 1$.

You'll need to `import` the `solve_ivp` function from `scipy`. Solve for the solution between $x=0$ and $x=30$ and pass the optional argument `t_eval=np.linspace(0,30,10000)` to `solve_ivp` to densely sample the result.

```python
from scipy.integrate import solve_ivp
rho=28
sigma=10
beta=8/3
def Lorenz(t,r):
    x,y,z = r
    return [(sigma*(y-x)),(x*(rho-z)-y),(x*y-beta*z)]
t=np.linspace(0,30,10000)
solLorenz=solve_ivp(Lorenz,[0,30],[1,1,1],t_eval=t)   
```

Assign your solution to a variable called `solLorenz` and use the following code to make a 3D plot of the solution

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.plot(solLorenz.y[0], solLorenz.y[1], solLorenz.y[2]);
```

(Q3b): The Lorenz ODEs are a classic example of a system of ODEs which exhibit chaos, i.e., small changes in the initial parameters will result in large changes to the solution at late times. To see this, solve the Lorenz system again with initial parameters $x(0) = 1 + 10^{-5}, y(0) = 1, z(0) = 1$, and plot both this solution and the previous solution on a new 3D plot.

```python
solLorenz2=solve_ivp(Lorenz,[0,30],[1+1e-5,1,1],t_eval=t)
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.plot(solLorenz.y[0], solLorenz.y[1], solLorenz.y[2],label='first solution');
ax.plot(solLorenz2.y[0], solLorenz2.y[1], solLorenz2.y[2],color='red',label='second solution');
plt.legend()
plt.show()
```

You should see that initially the two curves are together but later one they start to diverge. The Lorenz system was original developed as a simple model of atmospheric convection and already this simple system, much like real weather, displays chaotic behaviour.

You should find the systems with initial values $u(0) = \{0,1\}$ do not evolve as the right-hand side equals zero in these cases. These two values are called *equilibrium points*. You should observe that systems with $u(0) > 0$ should tend to $u=1$ for large $x$ (and away from $u=1$). For this reason we say that $u=1$ is a stable equilbrium and $u=0$ is an unstable equilbrium.

```python
def uprime(x,u):
    return (u*(1-u))
x=np.linspace(0,8,8000)
u1=solve_ivp(uprime,[0,8],[0],t_eval=x)
u2=solve_ivp(uprime,[0,8],[0.01],t_eval=x)
u3=solve_ivp(uprime,[0,8],[0.2],t_eval=x)
u4=solve_ivp(uprime,[0,8],[0.4],t_eval=x)
u5=solve_ivp(uprime,[0,8],[0.6],t_eval=x)
u6=solve_ivp(uprime,[0,8],[0.8],t_eval=x)
u7=solve_ivp(uprime,[0,8],[1],t_eval=x)
u8=solve_ivp(uprime,[0,8],[1.2],t_eval=x)
u9=solve_ivp(uprime,[0,8],[1.4],t_eval=x)
u10=solve_ivp(uprime,[0,8],[1.6],t_eval=x)
u11=solve_ivp(uprime,[0,8],[1.8],t_eval=x)
plt.plot(u1.t,u1.y[0],label='u(0)=0')
plt.plot(u2.t,u2.y[0],label='u(0)=0.01')
plt.plot(u3.t,u3.y[0],label='u(0)=0.2')
plt.plot(u4.t,u4.y[0],label='u(0)=0.4')
plt.plot(u5.t,u5.y[0],label='u(0)=0.6')
plt.plot(u6.t,u6.y[0],label='u(0)=0.8')
plt.plot(u7.t,u7.y[0],label='u(0)=1')
plt.plot(u8.t,u8.y[0],label='u(0)=1.2')
plt.plot(u9.t,u9.y[0],label='u(0)=1.4')
plt.plot(u10.t,u10.y[0],label='u(0)=1.6')
plt.plot(u11.t,u11.y[0],label='u(0)=1.8')
plt.xlabel("x")
plt.ylabel("u(x)")
plt.grid(True, 'both', 'both', linestyle='--')
plt.legend(fontsize='x-small')
plt.show()
```
