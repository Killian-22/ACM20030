# Question 1: numerical integration

Use the `SimpsonsIntegrate` function below when answer this question

```python
def SimpsonsIntegrate(f, a, b, N):
    dx = (b-a)/N
    xi = a
    i = 0
    area = 0
    while i < N:
        area += dx/6*(f(xi) + 4*f((2*xi+dx)/2) + f(xi+dx))
        xi += dx
        i+= 1
        
    return area
```

## Q1a [4 marks]

Numerically evaluate $$\int_0^1 \sin(x)\, dx$$

Compare your answer to the analytic result and use enough strips in the integration so that the relative error drops below `1e-10`, and show your result is at least this accurate.

```python
def f(x):
    return np.sin(x)
I = SimpsonsIntegrate(f,0,1,50)
print("Numerical integration yields " + str(I))
print("And the analytic solution is " + str(-np.cos(1)+1))
print("Their difference is " + str(abs(I-(-np.cos(1)+1)))) 
```

```python
true_integral = -np.cos(1)+1
errs = [abs(SimpsonsIntegrate(f,0,1,10)-true_integral),abs(SimpsonsIntegrate(f,0,1,100)-true_integral),abs(SimpsonsIntegrate(f,0,1,1000)-true_integral)]
steps = [10,100,1000]
x = np.arange(10.0,1001.0,1)
y = x**-4
plt.plot(x,y,label="$x^{-4}$ reference line")
plt.scatter(steps,errs,color='red',label="Errors from step numbers of 10;100;1000")
plt.xscale('log')
plt.yscale('log')
plt.ylabel('error')
plt.xlabel('steps')
plt.ylim(10**-16,10**-3)
plt.legend()
plt.show()
```

## Q2a [4 marks]

The dot product of two vectors $a$, $b$, each of length $n$ is given by

$$ a\cdot b = \sum_{i=1}^n a_i b_i$$

Two sample vectors and their dot product are given below

```python
a = np.array([4.5, 6.7, 6.7])
b = np.array([8.8, -4.5, 3.3])

print("The dot product of a and b is:", np.dot(a,b))
```

Complete the `DotProduct` function below to compute the dot product by using a loop (do not use `np.dot()` or a similar function). Test your answer on the vectors $a$ and $b$.

```python
def DotProduct(a,b):
    if(len(a)!=len(b)):
        print("The vector lengths must be the same!")
        return None
    dotprod = 0
    for i in range(0,len(a)):
        dotprod += a[i]*b[i]
    return dotprod
```

```python
print("Using the custom function, the dot product of a and b is " + str(DotProduct(a,b)))
```

## Q2b [6 marks]

In the lectures we looked at the $L^2$-norm. Another matrix norm is the Frobenius norm, $\lVert \cdot \rVert_F$ of a $m\times n$ matrix $A$ is defined as 

$$ \lVert A \rVert_F = \sqrt{\sum_{i=1}^m \sum_{j=1}^n |a_{ij}|^2 }$$

Complete the `FrobeniusNorm()` function below make it compute the Frobenius norm of a matrix using loops (don't use `la.norm`, or a similar function). 

```python
def FrobeniusNorm(A):
    m,n=A.shape[0],A.shape[1]
    norm = 0
    for i in range(0,m):
        for j in range(0,n):
            norm += (A[i,j])**2
    norm = np.sqrt(norm)
    return norm            
```

Test your function using the below matrix. The `la.norm(A)` function returns the value of the Frobenius norm. You can use this to compare against your answer.

```python
A = np.array([[1.4,-3.5,4.6],[4.5, 3.5, 6.6],[-7.7,5.5,6.6],[4.5,-4.5,4.3]])
```

```python
print("Using npla.norm, we find the Frobenius norm of the above array to be " + str(npla.norm(A)))
print("Using the FrobeniusNorm function, we find the Frobenius norm of the array to be "
      +str(FrobeniusNorm(A)))
```

## Q3b [2 marks]

As we discused in the lectures, one way to define the condition number of a matrix is given by 

$$\kappa(A) = \lVert A \lVert_2 \lVert A^{-1} \rVert_2 $$

Compute $\kappa(B)$ using this formula and where `B` is the matrix defined in Q3a. You may use `la.inv(B)` and `la.norm(B,2)` in this question.

Hint: you can check you result against `la.cond(B,2)`

```python
#Yes I copied these function definitions from my homework.
def npArrayType(A): 
    if (isinstance(A,np.ndarray)==False):
        print("Your object must have array-like structure")
        return None
    if (isinstance(A,np.matrix)):
        return 'matrix'
    else: 
        return 'ndarray'

def ConditionNumber(M,method='norm'):
    if (method=='norm'):
        if (npArrayType(M)=='matrix'):
            return (npla.norm(M,2) * npla.norm(M**-1,2))
        elif(npArrayType=='ndarray'):
            return (npla.norm(M,2) * npla.norm(npla.inv(M),2))
    elif(method=='eigs'):
        if (npArrayType(M)=='matrix'):
            MTM = (M.T)*M
        elif(npArrayType(M)=='ndarray'):
            MTM = (M.T)@M
        eigvals = npla.eig(MTM)[0]
        max_eig=np.max(eigvals)
        min_eig=np.min(eigvals)
        return np.sqrt(max_eig/min_eig)
```
